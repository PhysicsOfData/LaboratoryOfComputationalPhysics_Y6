```python
# FUNZIONI
def square(x):
    return x*x
def cube(x):
    return x**3
print(square(3))
print(cube(3))
# posso fare un dizionario di funzioni
funcs = {'square': square, 'cube': cube}

# ARGOMENTI
# l'argomento delle funzioni può essere di molti tipi

def modify(x):
    x[0] = 999 #prende il primo valore di x e lo cambia qualunque esso sia in 999
    return x
x = [1,2,3]
print(modify(x))
```

```python
# HOF (HIGHER ORDER FUNCTIONS)

# MAPS
# applicano una funzione ad una collezione di oggetti
x = list(map(square, range(5)))
print(x)

# FILTER
# posso filtrare determinati valori che voglio e applicargli la funzione
def even(x):
    return x % 2 == 0
y = list(filter(even, range(5)))
print(y)

# COMBINAZIONI
# posso combinare hofs
z = list(map(square, filter(even, range(5))))
print(z)

# REDUCE
from functools import reduce
def add(x, y):
    return x+y
reduce(add, [1,2,3,4,5]) #fa la sommatoria in questo caso anche se l'argomento contiene solo 2 input

# ZIP
# applica funzioni a liste o dizionari multipli prendendo il primo elemento di ogni lista e così via
a = [1,2,3]
b = [4,5,6]
c = [10,20,30]
for ai,bi,ci in zip(a,b,c):
    print(ai,bi,ci)
```

```python
# LAMBDA
# senza creare funzioni posso usare lambda per chiamare una funzione che fa un piccolo task in una lista 
suml = lambda x,y: x+y #nome funzione = lamda argomento: definizione
print(suml(3,4))
# posso chiamarla in due differenti modi
z = (lambda x,y : x+y)(3,4)
print(z)
for i in map (lambda x: x*x, range(5)): print(i)
```

```python
# FUNZIONE RECURSIVA
# vediamo ad esempio fibonacci con la recursive (ciclo if)
def fib1(x):
    if x==0 or x==1: #caso base
        return 1
    else: #caso recursivo
        return fib1(x-1) + fib1(x-2)
result1 = [fib1(i) for i in range(10)] #creo lista dei risultati
print(result1)

# vediamo il caso non recursivo (ciclo for)
def fib2(x):
    a, b = 0, 1
    for i in range (1, x+1):
        a, b = b, a+b
    return b
result2 = [fib2(i) for i in range(10)]
print(result2)

# possiamo controllare quanto tempo ci mette il computer per calcolare il risultato nei due modi
%timeit fib1(20)
%timeit fib2(20) #il secondo risulta più veloce
```

```python
# DECORATORI
# modi per decorare con commenti una volta che fai un'azione
def decorator(func): #definisco il decoratore con argomento la funzione
    def wrapper(): #definisco il wrapper
        print('Prima')
        func()
        print('Dopo')
    return wrapper #ritorno il wrapper

def lmao(): #definisco una qualsiasi funzione
    print('Lmao')

ciao = decorator(lmao)
print(ciao()) #chiamo

#posso anche chiamare il decorator con la @
@decorator
def asd():
    print('asd')
asd()
```
